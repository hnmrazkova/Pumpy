```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_clean_data = pd.read_csv('clean_data.csv')
```

```python
df_clean_data
```

```python
df_clean_data.drop(df_clean_data.columns[0], axis = 1, inplace = True)
```

```python
df_clean_data
```

```python
def feature2number(df, column_name):
    unique_values = set(df[column_name].tolist())
    transformer_dict = {}
    for index, value in enumerate(unique_values):
        transformer_dict[value] = index
    df[column_name] = df[column_name].apply(lambda y: transformer_dict[y])
    return df
```

```python
df_clean_data.dtypes
```

```python
df_clean_data.dtypes
```

```python
numeric_columns = ['status_group', 'amount_tsh','count_month', 
                   'gps_height', 'construction_year', 
                   'missing_construction_year'] 
columns_to_transform = [col for col in df_clean_data.columns if col not in numeric_columns]
for column in columns_to_transform: 
    feature2number(df_clean_data, column)
  
```

```python
df_train_test = df_clean_data[df_clean_data['status_group'] != 3]
                                                           
```

```python
df_train_test.status_group.value_counts()
```

```python
df_to_predict = df_clean_data[df_clean_data['status_group'] == 3]
```

```python
df_train_test.to_csv('train_test_data.csv')
df_to_predict.to_csv('predict_data.csv')
```

```python
Xtrain = df_train_test.drop('status_group', axis = 1)
```

```python
ytrain = df_train_test['status_group']
```

```python
ytrain.value_counts()
```

```python
X_learn, X_test, y_learn, y_test = train_test_split(Xtrain, ytrain, test_size = 0.3)
```

```python
rfc = RandomForestClassifier(n_estimators = 100)
```

```python
rfc.fit(X_learn, y_learn)
```

```python
rfc.score(X_learn, y_learn)
```

```python
rfc.score(X_test, y_test)
```

```python
predicted_results = rfc.predict(df_to_predict.drop(['status_group'], axis = 1))
```

```python
predicted_results.size
```

```python
from sklearn.metrics import classification_report
```

```python
predicted = rfc.predict(X_test)
```

```python
report = classification_report(y_test, predicted)
```

```python
target_names = ['non functional 0', 'functional needs repair 1', 'functional 2']
```

```python
report = classification_report(y_test, predicted, target_names = target_names)
```
